# ***Especificaciones generales del diseño***

### Dimensiones del escritorio:
- **Ancho**: 80 cm
- **Profundidad**: 50 cm
- **Altura**: 75 cm

### Descripción general:
El escritorio cuenta con una superficie de trabajo plana y un respaldo elevado para evitar que los objetos caigan en la parte trasera. La estructura derecha del escritorio ofrece almacenamiento funcional con:
- Un compartimento superior para la caja del controlador del softbot (dimensiones: 23 cm x 33 cm x 13 cm).
- Dos estantes abiertos para almacenamiento adicional.

### Finalidad:
Está diseñado para satisfacer necesidades de trabajo y almacenamiento eficiente, adecuado para una estación de trabajo establecida. Las características incluyen:
- Superficie para colocar y manejar una computadora.
- Espacio para la caja del controlador del softbot.
- Dos espacios adicionales para almacenamiento.

# **Materiales y acabados**

### Material principal:
El material recomendado es MDF de 19 mm (¾ de pulgada) de grosor para rigidez y durabilidad. Alternativas incluyen:
- Madera maciza de 25 mm (1 pulgada).
- Contrachapado de 21 mm (⅞ de pulgada).
Estos materiales son resistentes, duraderos y ofrecen un acabado suave que puede ser barnizado o pintado.

### Acabado:
- **Barniz resistente a la humedad y al rayado**:
  - Proporciona resistencia a la humedad y al rayado, mejorando la estética con un acabado suave y brillante.
- **Laminado**:
  - Ofrece durabilidad, resistencia al desgaste y un mantenimiento sencillo.

# **Compartimentos y funcionalidad del almacenamiento**

- **Superficie superior**: Espacio suficiente para una laptop y otros objetos de trabajo, con respaldo elevado.
- **Compartimento para la caja del controlador**: Ubicación segura y accesible.
- **Estantes abiertos inferiores**: Ideales para documentos, accesorios o herramientas pequeñas; de fácil acceso.

# **Detalles del ensamble y uniones**

- **Tornillos y conectores metálicos (soportes en L)**: Alta resistencia y facilidad de montaje.
- **Adhesivo resistente**: Refuerzo estructural en puntos estratégicos.
- **Uniones de ranura y lengüeta**: Ensamble preciso y sólido.

### Detalles del montaje:
Los tornillos y soportes se colocan estratégicamente para evitar movimiento y asegurar rigidez estructural.

# **Aspectos de fabricación**

### Corte y preparación de las piezas:
Las piezas deben cortarse con precisión y lijarse para evitar astillas y lograr un buen acabado.

### Montaje:
Se aplica adhesivo en las uniones principales antes de fijar con tornillos para mejorar la durabilidad.

# **Ergonomía**

- Altura ajustada para comodidad.
- Espacios de almacenamiento diseñados para un fácil acceso.

# **Reacondicionamiento del robot**

## Antes de la limpieza:
Previo al proceso de limpieza del robot y la mesa de trabajo, se realizó una inspección para identificar la presencia de polvo, residuos y otros componentes no deseados que pudieran comprometer la operación y seguridad del sistema robótico. Esta revisión permitió determinar las áreas críticas que requerían mayor atención, como juntas, sensores, superficies de contacto y componentes delicados. 

Además, se aseguraron las condiciones de seguridad necesarias para prevenir cualquier posible daño al equipo durante la limpieza. Como podemos apreciar en las imágenes siguientes, en la mesa de trabajo se encontraron papeles, polvo, tornillos y cosas pegajosas. En la parte del robot únicamente se encontró polvo.

# **Conclusión**

En conclusión, este escritorio ofrece una solución funcional y optimizada para el trabajo y almacenamiento. Su diseño eficiente proporciona gran accesibilidad a cada compartimento. Las uniones y materiales sugeridos aseguran un producto de calidad, estable, duradero y fácil de ensamblar, contribuyendo a su practicidad y utilidad. 

Este escritorio está diseñado para permitir el uso cómodo de una computadora y proporciona un espacio necesario para manejar de manera ergonómica y efectiva. Además, incluye un espacio dedicado para alojar el controlador del robot, ofreciendo suficiente lugar para los cables, evitando que se doblen o rompan. 

El diseño maximiza la funcionalidad, asegurando comodidad y eficiencia para cumplir con cada una de las restricciones iniciales.
