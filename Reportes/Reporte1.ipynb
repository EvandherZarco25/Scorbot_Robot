## Resumen
<div style="text-align: justify;">
El manipulador Scorbot requiere realizar un proceso de "Home" para sincronizar sus articulaciones mediante encoders que envían pulsos, los cuales permiten conocer la posición y velocidad del brazo. Se presentan problemas en los ejes, especialmente en el eje 5, afectando el movimiento de la muñeca. Las dificultades incluyen fallos de ensamblaje y acumulación de óxido y polvo, impactando el rendimiento impidiendo llegar a "Home".
</div>

## Descripcion de brazo
<div style="text-align: justify;">
El Scorbot cuenta con múltiples ejes controlados por encoders, incluyendo el eje 5 que permite el giro de la muñeca. Está diseñado para operar mediante un sistema de retroalimentación sencillo de pulsos, adecuado para el control de posición y velocidad de sus articulaciones.
</div>



## Evaluacion de estado de componentes
<div style="text-align: justify;">
El sistema presenta limitaciones en la retroalimentación, ya que el uso de pulsos no garantiza una precisión ideal en el control de las articulaciones. Además, el eje 5 sufre un mal funcionamiento debido a la instalación incorrecta de una cámara, lo que interfiere con su sensor de posición. La presencia de óxido y polvo también se observa en el robot, afectando el desempeño general.
</div>


## Registros de daños

- Eje 5: Fallo en el movimiento de la muñeca causado por la instalación de una cámara.
- Encoders y articulaciones: Errores en la posición debido a la acumulación de polvo y posibles obstrucciones en el proceso de "Home".
- Óxido: La presencia de óxido en componentes metálicos disminuye la eficiencia y precisión.

## Recomendaciones de mantenimiento

- Limpieza: Realizar limpieza con pistola de aire en las articulaciones y zonas de movimiento.
- Revisión del Eje 5: Verificar la instalación de componentes adicionales y corregir el ensamblaje del sensor de posición.
- Comparación de Diagnóstico: Comparar el estado actual con otros Scorbots en buen estado para un análisis detallado de problemas.
- Regreso a "Home": Después de cada uso, llevar el brazo a la posición "Home" para garantizar una mejor inicialización en futuros encendidos.
